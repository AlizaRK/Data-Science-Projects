```python
import pandas as pd
import datetime
pd.options.mode.chained_assignment = None
import numpy as np
```

```python
data = pd.read_csv("580SurveyCleanup_formatted.csv")
```

```python
data = data.drop(columns='Unnamed: 0') # repeated column
```

### Columns with missing values

```python
data.columns[data.isnull().any()]
```

### Firstly, replacing all null values with ''

```python
data = data.fillna('')
data
data['Section'] = data['Section'].astype('string')
```

### It is important to fix specific values of data (cleaning) before filling in the missing values.

### Cleaning the data

#### Columns that require cleaning:

```python
data[['Age (years)', 'Height (Inches)', 'Country of Citizenship', 'Undergraduate Degree', 
      'What is your goal for learning about data analytics?']]
```

#### Age (years)
Two values need to be fixed:
1. Value in row 17: 2 3 -> 23
2. Value in row 32: age needs to be calculated from 1993.

For the second point, I will calculate data from 1993 to 2022.

```python
# Correct 1
data['Age (years)'] = data['Age (years)'].replace('2 3', '23').astype('int')
# Correct 2
data['Age (years)'] = data['Age (years)'].replace(1993, str(2022-1993)).astype('int') # do better
# Final conversion to integer
data['Age (years)'] = data['Age (years)'].astype('int')
data['Age (years)']
```

#### Height (inches)
Following changes are required:
1. In row 7, 5.2 value is not inches, possibly in feet that will be converted to inches. 
2. In row 27, the value "6I" is a typo. It might possibly mean 61. 
3. In row 32, the value "7O" is a typo. It might possibly mean 70. 

```python
# Correction 1:
new_val = int(5.2 * 12)
data['Height (Inches)'] = data['Height (Inches)'].replace('5.2', str(new_val))

# Correction 2:
data['Height (Inches)'] = data['Height (Inches)'].replace('6l', '61')

# Correction 3:
data['Height (Inches)'] = data['Height (Inches)'].replace('7O', '70')
data[['Age (years)', 'Height (Inches)']]
```

#### Country of Citizenship
Following changes are needed:
1. All the values need to have same string pattern for this case, (title form)
2. In row 32, dual citizenship is mentioned. For this, a semicolon would be removed and the entries will be separated by a coma. 
3. All USA or US (line 4) will be converted to Unites States.
4. Removing all trailing spaces. 

```python
# Correction 3
data.loc[data['Country of Citizenship'].str.contains("US"), 'Country of Citizenship'] = 'United States'

# # Correction 2
data['Country of Citizenship'][32] = data['Country of Citizenship'][32].replace(' ; ', ' ') 

# Correction 1
data['Country of Citizenship'] = pd.Series(data['Country of Citizenship']).str.title()

# Correction 4
data['Country of Citizenship'] = pd.Series(data['Country of Citizenship']).str.strip()

data[['Age (years)', 'Height (Inches)', 'Country of Citizenship']]
```

#### Undergraduate Degree
Following tasks are performed:
1. Making all the strings title.
2. Removing trailing spaces.

```python
# Correction 1
data['Undergraduate Degree'] = pd.Series(data['Undergraduate Degree']).str.title()

# Correction 2
data['Undergraduate Degree'] = pd.Series(data['Undergraduate Degree']).str.strip()

# Correction 3
data.loc[data['Undergraduate Degree'].str[:3] == "Eco", 'Undergraduate Degree'] = 'Economics'
data[['Age (years)', 'Height (Inches)', 'Country of Citizenship', 'Undergraduate Degree']]
```

#### Analytics Goal
Following are the changes required:
1. Trailing spaces needs to be removed.

```python
# Correction 1
data['What is your goal for learning about data analytics?'] = pd.Series(
    data['What is your goal for learning about data analytics?']).str.strip()
data['What is your goal for learning about data analytics?'] = pd.Series(
    data['What is your goal for learning about data analytics?']).str.title()
data[['Age (years)', 'Height (Inches)', 'Country of Citizenship', 'Undergraduate Degree', 
      'What is your goal for learning about data analytics?']]
```

### Filling the missing values

#### Access Code and Email Address
Since both these columns have all null values, there is no information to be preserved about these columns. Therefore, they can be dropped.

```python
data = data.drop(columns=['Access Code', 'Email Address'], axis=1)
data
```

#### Region
For this column, we will take a mode of the values to fill in as we can see certain values appearing a lot of times. This also seems like a categorical data.

```python
data['Region'] = data['Region'].replace('', data['Region'].mode()[0])
data['Region']
```

#### City / Town / District
Again, here as well taking the mode of the data to fill in the empty values. 

```python
data['City / Town / District'] = data['City / Town / District'].replace('', data['City / Town / District'].mode()[0])
data[['Region','City / Town / District']]
```

#### Height (Inches)
Taking the mean height value of the data and filling the missing values with that.

```python
data['Height (Inches)'] = data['Height (Inches)'].replace('', '0')
data['Height (Inches)'] = data['Height (Inches)'].astype('int')

data.loc[data['Height (Inches)'] == 0, 'Height (Inches)'] = data['Height (Inches)'].mean(0, skipna=True)
data['Height (Inches)'] = data['Height (Inches)'].astype('int')
data[['Region','City / Town / District', 'Height (Inches)']]
```

#### Country of Citizenship
For this we will again take the mode as it is a nominal data.

```python
data['Country of Citizenship'] = data['Country of Citizenship'].replace('', data['Country of Citizenship'].mode()[0])
data[['Region','City / Town / District', 'Height (Inches)', 'Country of Citizenship']]
```

#### Undergraduate Degree
This is also a Nominal data so we will take the mode to fill out missing values.

```python
data['Undergraduate Degree'] = data['Undergraduate Degree'].replace('', data['Undergraduate Degree'].mode()[0])
data[['Region','City / Town / District', 'Height (Inches)', 'Country of Citizenship', 'Undergraduate Degree']]
```

#### Expected Graduation date from Mason MS program
Setting the missing values to today's date as there is no mode for this column.

```python
data['Expected Graduation date from Mason MS program?'] = data['Expected Graduation date from Mason MS program?'].replace('', pd.to_datetime('today').date())
data['Expected Graduation date from Mason MS program?']
```

#### What is your goal for learning about data analytics?
Filling the missing values with "No Response" as it is an open-ended question and no statistical method would be appropriate to fill in the values. 

```python
data['What is your goal for learning about data analytics?'] = data['What is your goal for learning about data analytics?'].replace('', 'No Response')
data['What is your goal for learning about data analytics?']
```

#### Employment
It is a Nominal data, we will use mode to fill in the values. 

```python
data['Employment'] = data['Employment'].replace('', data['Employment'].mode()[0])
data['Employment']
```

#### Python Proficiency
Taking the mode and filling in the missing values. Reason is same as above.

#### Type of Laptop
Taking the mode and filling in the missing values. Reason is same as above.

```python
data['Type of Laptop'] = data['Type of Laptop'].replace('', data['Type of Laptop'].mode()[0])
data['Type of Laptop']
```

#### R Proficiency
Taking the mode and filling in the missing values. Reason is same as above.

```python
data['R Proficiency'] = data['R Proficiency'].replace('', data['R Proficiency'].mode()[0])
data['R Proficiency']
```

#### SQL Proficiency
Taking the mode and filling in the missing values. Reason is same as above.

```python
data['SQL Proficiency'] = data['SQL Proficiency'].replace('', data['SQL Proficiency'].mode()[0])
data['SQL Proficiency']
```
